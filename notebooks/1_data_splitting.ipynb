```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Train / Valid / Test 
path = "../data/ArchiveII_max_length_128.csv"
df = pd.read_csv(path)
df.structure.sort_values()
```

```python
# Filter out sequences without structures
print(df.shape)
for i in [2309, 2002,479, 312]:
    print(df.iloc[i].structure)
    df.drop(index=i, inplace=True)
print(df.shape)
```

```python

train_df, rest_df = train_test_split(
    df, train_size=0.70, random_state=42, stratify=df['fam']
)

val_df, test_df = train_test_split(
    rest_df, train_size=0.50, random_state=42, stratify=rest_df['fam']
)

train_df.shape, val_df.shape, test_df.shape
```

```python
save_path = "../data/ArchiveII_128_random_split/"

train_df.to_csv(save_path + "train.csv", index=False)
val_df.to_csv(save_path + "val.csv", index=False)
test_df.to_csv(save_path + "test.csv", index=False)
```
