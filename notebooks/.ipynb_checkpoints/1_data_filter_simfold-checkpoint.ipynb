```python
import pandas as pd

rows = []

df_filt = {}

for sim, df_s in sim_df.items():
    df_filt[sim] = df_s[df_s["id"].isin(valid_ids)].reset_index(drop=True)

    # Conteos FULL
    full_counts = (
        df_s.query("fold_number == 0")["partition"]
        .value_counts()
        .to_dict()
    )

    # Conteos FILTRADO (max128)
    filt_counts = (
        df_filt[sim].query("fold_number == 0")["partition"]
        .value_counts()
        .to_dict()
    )

    # Guardamos en formato largo
    for part, n in full_counts.items():
        rows.append({
            "sim": sim,
            "dataset": "full",
            "partition": part,
            "count": n
        })

    for part, n in filt_counts.items():
        rows.append({
            "sim": sim,
            "dataset": "max128",
            "partition": part,
            "count": n
        })

# DataFrame final
counts_df = pd.DataFrame(rows)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(data=counts_df, x="sim", y="count", hue="dataset", ci=None)
plt.title("Cantidad de muestras por simulación")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=counts_df, x="partition", y="count", hue="dataset", ci=None)
plt.title("Distribución de particiones")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

g = sns.catplot(
    data=counts_df,
    kind="bar",
    x="partition",
    y="count",
    hue="dataset",
    col="sim",
    col_wrap=4,
    height=4,
    sharey=False
)

g.set_titles("{col_name}")
g.set_axis_labels("Partición", "Cantidad de muestras")

g.fig.subplots_adjust(top=0.9, right=0.85)  
g.add_legend(title="Dataset")
g._legend.set_bbox_to_anchor((1.02, 0.5))   

plt.show()
```
